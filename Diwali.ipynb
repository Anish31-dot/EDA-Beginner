```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
df
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],inplace=True,axis=1)
```

```python
df.head(5)
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.describe()
```

```python
ax=sns.countplot(data=df,x='Gender',palette='viridis')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_count = df['Gender'].value_counts()
gender_count
```

```python
plt.pie(x=gender_count.values,autopct='%.1f%%',labels=gender_count.index)
plt.show()
```

```python
sales = df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
sales
```

```python
ax=sns.barplot(x=sales.index,y=sales.values,palette='viridis')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data=df,x='Occupation',palette='viridis')
```

```python
occupation = df['Occupation'].value_counts().sort_values(ascending=False)
```

```python
occupation
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=occupation.index,y=occupation.values,palette='viridis')
```

```python
sns.boxplot(data=df['Age'])
```

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr
df1=df[(df['Age']>= lower) & (df['Age']<= upper)]
```

```python
sns.boxplot(df1['Age'])
```

```python
df1
```

```python
productcat_Amount = df1.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(5)
productcat_Amount
```

```python
# sns.barplot(x=productcat_Amount.index,y=productcat_Amount.values,palette='viridis')
# plt.show()
plt.pie(labels=productcat_Amount.index,x=productcat_Amount,autopct='%.1f%%')
#plt.savefig('Product Category By Amount.jpg')
plt.show()
```
