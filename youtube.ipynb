```python
sns.boxplot(data=df)
```

```python
# IQR method to remove outliers
q1 = df['view_count'].quantile(0.25)
q3 = df['view_count'].quantile(0.75)
IQR = q3 - q1
```

```python
IQR
```

```python
lower_bound = q1 - 1.5*IQR
upper_bound = q3 + 1.5*IQR
df1 = df[(df['view_count']>= lower_bound) & (df['view_count']<= upper_bound)]
df1
```

```python
top5_channel_duration = df1.groupby('channel')['duration'].sum().sort_values(ascending = False).head(5)
top5_channel_duration
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top5_channel_duration.index,y=top5_channel_duration.values,palette='viridis')
plt.title('Top 5 Channels by Duration')
plt.xlabel('Channel')
plt.ylabel('Total Duration')
plt.show()
```

```python
sns.countplot(x='live_status',data=df1)
```

```python
df1.drop(['live_status'],axis=1,inplace=True)
```

```python
top5_categories_duration = df1.groupby('categories')['duration'].sum().sort_values(ascending=False)
top5_categories_duration
```

```python
# plt.pie(top5_categories_duration,labels=top5_categories_duration.index,explode=[0.05]*len(top5_categories_duration),autopct='%.1f%%')
# plt.title('Top 5 Categories by Duration')
# plt.show()
import warnings
warnings.filterwarnings("ignore")
fig,ax = plt.subplots(figsize=(10,6))
sns.barplot(x=top5_categories_duration.index,y=top5_categories_duration.values,palette='viridis',ax=ax)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Top 5 Categories by Duration')
plt.show()
```
